```python
def clean_string(text):
    text = regex.sub(r"-\n(?=[a-z])", r"", text) # ver-beteren -> verbeteren
    text = regex.sub(r"-\n(?=[A-Z])", r"-", text) # Bert-Jan -> Bert-Jan
    text = regex.sub(r"\n", " ", text) # remove all remaining newlines
    text = regex.sub(r"\s+", " ", text) # change all multiple whitespace characters to a single space
    return text
```

```python
def create_jsonl(years, data, prompt):
    lines = []
    # create test sample size 10
    #data = data.sample(n=10, random_state=22)
    for nr, idx in enumerate(data.index):
            messages=[
                    {"role": "system", "content": f"""{prompt}"""},
                    {"role": "user", "content": f"""{data['TEXT_GPT'].loc[idx]}"""}, 
            ]
            line = {"custom_id": data['ID'].loc[idx], "method": "POST", "url": "/v1/chat/completions", "body": {"model": "gpt-4-0125-preview", "temperature":0, "response_format":{"type":"json_object"}, "messages": messages}}
            lines.append(line)
    with jsonlines.open(f'data_gpt/{years}.jsonl', mode='w') as writer:
        for line in lines:
            writer.write(line)
```
