```python
from jiwer import wer, cer
import nltk

def calc_cer(df, p, column_name_2):
    if df['name'] == '' and df[column_name_2] == '' or df['name'] == ' ' and df[column_name_2] == ' ':
        return None
    elif df['name'] == '' or df[column_name_2] == '' or df['name'] == ' ' or df[column_name_2] == ' ':
        return 1
    else:
        cer_1 = cer(df['name'],df[column_name_2])
        cer_2 = cer(df['name_reversed'],df[column_name_2])
        cer_3 = cer(df['{}_fname'.format(p)], df[column_name_2])
        return min(cer_1,cer_2,1,cer_3)
    
def calc_levst(df, p, column_name_2):
    if df['name'] == '' and df[column_name_2] == '' or df['name'] == ' ' and df[column_name_2] == ' ':
        return None
    else:
        lvst_1 = nltk.edit_distance(df['name'],df[column_name_2])
        lvst_2 = nltk.edit_distance(df['name_reversed'],df[column_name_2])
        lvst_3 = nltk.edit_distance(df['{}_fname'.format(p)], df[column_name_2])
        return min(lvst_1,lvst_2,lvst_3)
```
